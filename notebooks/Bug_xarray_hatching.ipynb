# Hatching density can not be controlled using xarray

```python
import xarray as xr
import matplotlib.pyplot as plt

air = xr.tutorial.open_dataset('air_temperature').air
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
air.isel(time=0).plot.contourf(ax=ax[0], hatches=['/'])
air.isel(time=0).plot.contourf(ax=ax[1], hatches=['////'])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].contourf(air.isel(time=0), hatches=['/'])
ax[1].contourf(air.isel(time=1), hatches=['////'])
```
